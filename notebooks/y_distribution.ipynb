Distribution of Seebeck in crysatl polyhedra dataset

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
poly = pd.read_csv(
    f"/Users/alina/PycharmProjects/ml-selection/data/processed_data/large_poly_descriptor.csv",
)
seebeck = pd.read_json(
    "/Users/alina/PycharmProjects/ml-selection/data/raw_data/median_seebeck.json", orient='split',
)
dataset = pd.merge(seebeck, poly, on="phase_id", how="inner").drop(columns=['phase_id', 'Formula'])
```

```python
plt.plot([i for i in range(len(dataset))], dataset['Seebeck coefficient'], 'o')
plt.show()
```

```python
plt.figure(figsize=(14, 7))
plt.hist(dataset['Seebeck coefficient'], bins=120, color='skyblue', edgecolor='black')
plt.xlabel('Seebeck Coefficient')
plt.ylabel('Frequency')
plt.show()
```
