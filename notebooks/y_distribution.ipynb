Distribution of Seebeck in crysatl polyhedra dataset

```python
import polars as pl
import matplotlib.pyplot as plt
import numpy as np
```

```python
dataset = pl.read_json('/root/projects/ml-selection/data/raw_mpds/total.json')
print(dataset)
```

```python
plt.plot([i for i in range(len(dataset))], dataset['Seebeck coefficient'], 'o')
plt.show()
```

```python
plt.figure(figsize=(14, 7))
plt.hist(dataset['Seebeck coefficient'], bins=120, color='skyblue', edgecolor='black')
plt.xlabel('Seebeck Coefficient')
plt.ylabel('Frequency')
plt.show()
```
